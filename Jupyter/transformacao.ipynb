```python
import pandas as pd
import pandera as pa
from datetime import date, time, datetime, timedelta
```

```python
valores_ausentes = ['**','####','###!','****','*****','NULL','***']
df = pd.read_csv("ocorrencia.csv", 
                 sep = ";", 
                 parse_dates = ['ocorrencia_dia'], 
                 dayfirst = True, 
                 na_values = valores_ausentes )
#na_values poderia ser usado como -> df.replace(['**','####','###!','****','*****','NULL'], pd.NA, inplace = True)
df.drop(['codigo_ocorrencia1',
         'codigo_ocorrencia3',
         'codigo_ocorrencia4',
         'ocorrencia_latitude',
         'ocorrencia_longitude',
         'codigo_ocorrencia2',
         'ocorrencia_pais',
         'investigacao_aeronave_liberada',
         'investigacao_status',
         'divulgacao_relatorio_numero',
         'divulgacao_relatorio_publicado',
         'divulgacao_dia_publicacao',
         "ocorrencia_hora",
         'total_aeronaves_envolvidas',
         'ocorrencia_saida_pista'], inplace=True, axis=1)
df.drop_duplicates(inplace=True)
```

```python
schema = pa.DataFrameSchema(
        columns = {
            "Não_necessária":pa.Column(pa.Int, required = False),     
            "ocorrencia_classificacao":pa.Column(pa.String),
            "ocorrencia_cidade":pa.Column(pa.String),
            "ocorrencia_uf":pa.Column(pa.String, pa.Check.str_length(2,2), nullable = True),
            "ocorrencia_aerodromo":pa.Column(pa.String, nullable = True),
            "codigo_ocorrencia":pa.Column(pa.Int),
            "ocorrencia_dia":pa.Column(pa.DateTime),
            "total_recomendacoes":pa.Column(pa.Int),
        }
)
```

```python
schema.validate(df)
```

```python
if(df.codigo_ocorrencia.is_unique):
    df.set_index('codigo_ocorrencia', inplace = True)
```

```python
data_atual = datetime.now()
data_atual_str = data_atual.strftime('%d-%m-%Y_%H:%M:%S')
df.to_excel('/home/ltk/DEV/BootCamp-Python/Data_engineer/Jupyter/{}.xlsx'.format(data_atual_str))
```
